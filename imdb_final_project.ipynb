# Q5. How do we clean the 'duration' column (e.g., '175 min' -> 175)?

```python
df["duration"] = df["duration"].str.replace("min","").str.replace(",","").astype(int)
```

# Q6. How do we clean the 'votes' column?

```python
df["votes"]=df["votes"].str.replace(",","").astype(int)
```

# Q7. How do we clean the 'gross_income' column (e.g., '175 min' -> 175)?

```python
df["gross_income"]=pd.to_numeric(df["gross_income"].str.replace(",",""),errors= "coerce")
```

```python
#info about the dataset

df.info()
```

```python
df.dropna(subset=["rating"],inplace=True)
```

```python
df["rating"].fillna(df["rating"].mean(),inplace = True)
```

# *Q15.How many movies are made each year*

```python
qw = df["year"].value_counts().sort_index()
plt.figure(figsize=(10,6))

sns.lineplot(x=qw.index , y= qw.values, color="red", marker="o", linestyle="--",label="movies")

plt.title("Number of Movies per Year")
plt.ylabel("Count")
plt.xlabel("Year")
plt.legend()
plt.show()
```

# Q18. Are ratings improving over the years?

```python
yearly_ratings = df.groupby("year")["rating"].mean()

plt.figure(figsize=(12, 6))
plt.plot(yearly_ratings.index, yearly_ratings.values, marker='o', linestyle='-', color='b')


plt.title('Average Ratings Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.show()
```

# Top 10 Genres

```python
y = df["genre"].str.split(",").explode("genre")
gen = df.groupby(y)["rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(x = gen.index, y = gen.values, palette="pastel")

plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.show()
```
