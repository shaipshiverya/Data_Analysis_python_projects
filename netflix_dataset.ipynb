# Netflix Content Analysis: An Exploratory Data Study

Netflix has grown to become one of the world's leading streaming platforms,
hosting thousands of titles spanning movies, TV shows, and original productions
across the globe. This project presents an exploratory analysis of Netflix's
content library, aiming to uncover patterns and trends within its catalog.

Through this analysis, we explore key dimensions of the platform's content
strategy, including the distribution between movies and series, the geographic
origins of productions, genre popularity, content ratings, release year trends,
and the most prominent directors and actors featured on the platform.



Import the dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dat file

```python
df = pd.read_csv('/content/netflix dataset.zip')
```

```python
# First five rows

df.head()
```

```python
#Sample of the data file

df.sample(10)
```

```python
#Statistical distribution of the data

df.describe()
```

```python
df.info()
```

```python
# Looking for the null values in the dataset

df.isna().sum()
```

# Imputing the values in the null columns

```python
df['director']=df['director'].fillna('unknown')
df['cast']=df['cast'].fillna('unknown')
df['country']=df['country'].fillna('unknown')
df['rating']=df['rating'].fillna('not rated')
```

```python
# check for null values again

df.isna().sum()
```

```python
# number of rows and columns

df.shape
```

Dropping the null values

```python
df.dropna(subset = ['duration'],inplace= True)
```

```python
df.drop_duplicates(inplace= True)
```

Remove the strings  from  the duration column and make it numerical


```python
df["duration"]=df['duration'].str.replace('min',"").str.replace("Seasons","").str.replace("Season","")
```

```python
# Standardising the date column for further analysis

df["date_added"]  = pd.to_datetime(df["date_added"],format = "mixed")
df["date_added"]
```

```python
df.sample(5)
```

# **EDA for the netflix dataset**

# **Movies and Series distribution in the app**

```python
sns.countplot(data =df , x ='type',color = 'maroon')


plt.title("Movies vs TV Shows on Netflix")
```

# Country making most of the content in netflix

```python
df['country'].value_counts().head(20)
```

```python
sns.countplot(data=df, y ='country', order=df['country'].value_counts().head(12).index, palette="viridis")
```

# Most Popular **Genres**

```python
genre = df['listed_in'].str.split(",").explode("listed_in")
genre.value_counts(ascending= False)
```

```python
sns.countplot( y = genre, order = genre .value_counts() .head(20) .index)
```

# **RATING DISTRIBUTION **

```python
df['rating'].value_counts()
```

```python
sns.countplot (data= df , y = 'rating', order= df['rating'] .value_counts() .index)
```

# **Release Year Analysis**

```python
rel_year = df['release_year'].value_counts(ascending= False).head(10)
rel_year
```

```python
plt.figure(figsize=(8,4))

sns.lineplot(x = rel_year.index, y = rel_year.values, color = 'green', linestyle=':', linewidth=4)

plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.title('Number of Titles Released by Year')

plt.show()
```
